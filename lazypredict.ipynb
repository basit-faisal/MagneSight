Using LazyPredict (Credible Library) to quickly compare ML Models for the predictions we want

```python
import lazypredict 
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyRegressor
import pandas as pd

```

Loading Data

```python
data = pd.read_csv('supermarket_sales.csv')
X_train, X_test, y_train, y_test = train_test_split(data[['cogs', 'Quantity', 'Gender']], data['gross income'], test_size=0.2, random_state=42)
```

Define and build lazy Regressor

```python
reg = LazyRegressor(verbose=0,ignore_warnings=True,custom_metric=None)
models,predictions = reg.fit(X_train,X_test,y_train,y_test)
```
